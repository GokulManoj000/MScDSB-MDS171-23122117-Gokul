```python
# restaurant
# customers -> orders(Menu items)
# collecting orders
# billing
# print bill
# update bill
# search for an order
import random

class restaurant:
    def __init__(self):
        self.menu= {
            "Rice":40,
            "Idli":20,
            "Roti":10
        }


        self.order={}


    def printmenu(self):
        print("*"*30)
        print("The current options in the menu are:")
        print("*"*30)
        print(self.menu)
        print("*"*30)

    def generateorderid(self):
        orderid=''
        for i in range(1,6,1):
            orderid+=str(random.randint(i,10))
        return orderid



    def collectorder(self):
        self.printmenu()# print menu
        phone=input("Enter the customer phone number:")
        order={}
        while True:
            a=False #to check item already exist
            item=input("Enter the item to be ordered:").title()
            qty=int(input("Enter the quantity:"))
            for i in order.keys():
                if i==item:
                    order[item]=order[item]+qty
                    a=True
                    break
            if a==False:
                order[item]=qty
            choice=input("Do you want tot complete the order? 'Y'").strip().upper()
            if choice=='Y':
                break
        orderid=self.generateorderid()
        self.order[orderid]={"Phone":phone,"Order":order}
        print(self.order)

    def billing(self):
        total = 0
        totalEarn = 0
        for orderid in self.order:
            print(self.order[orderid]['Order'])
            for key in self.order[orderid]['Order']:
                print(key,self.order[orderid]['Order'][key],self.menu[key])
                #print(self.order[orderid]['Order'][key]) #qty
                #print(self.menu[key]) #unit price
                price = self.order[orderid]['Order'][key] * self.menu[key]
                total = total+price
                totalEarn = totalEarn+total
                print("price =",price)
            print("total =",total)
        print("Earning of today is: ",totalEarn,"Rupees")
        
        

cftb=restaurant()
cftb.collectorder()
cftb.billing()
```
