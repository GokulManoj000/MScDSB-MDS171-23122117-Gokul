```python
import random
item=["Apple","Orange","Mango","Banana","Grapes","Guava","Jackfruit","Pomegranate","Coconut","Cherry"]
quantity=[4,7,5,8,14,6,9,3,10,12]
price=[100,125,50,80,55,90,75,95,60,70]
file=open("Lab06.csv","w+")
for a in range(0,100):
        item_index=random.choice(item)
        qty_index=random.choice(quantity)
        price_index=random.choice(price)
        print(a+1,item_index,qty_index,price_index)
        file.write(str(item_index)+","+str(qty_index)+","+str(price_index)+","+"\n")
file.close()

          

```
